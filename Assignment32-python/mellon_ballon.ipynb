```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import ListedColormap
from mpl_toolkits import mplot3d
```

```python
N  = 200
ballon_width = np.random.normal(4,1,N) 
ballon_length = np.random.normal(8,1,N)
ballon_weight = np.random.normal(2,1,N)

mellon_width = np.random.normal(10,1,N) 
mellon_length = np.random.normal(15,1,N)
mellon_weight = np.random.normal(20,1,N)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(ballon_width,ballon_length,ballon_weight,c="red",label='ballon')
ax.scatter3D(mellon_width,mellon_length,mellon_weight,c="yellow",label='ballon')
ax.set_xlabel("width")
ax.set_ylabel("length")
ax.set_zlabel("weight")
```
